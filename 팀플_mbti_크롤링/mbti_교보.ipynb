```python
import requests
from bs4 import BeautifulSoup
import openpyxl
import pandas as pd
```

```python
# 교보문고

pageNum =1
for i in range(1, pageNum+1):
    response = requests.get('https://search.kyobobook.co.kr/web/search?vPstrKeyWord=mbti&searchPcondition=1&searchCategory=%EA%B5%AD%EB%82%B4%EB%8F%84%EC%84%9C@KORBOOK&collName=KORBOOK&from_CollName=%EA%B5%AD%EB%82%B4%EB%8F%84%EC%84%9C@KORBOOK&searchOrder=0&vPstrTab=PRODUCT&from_coll=KORBOOK&currentPage=1&orderClick=LAH')
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    title = soup.select('div.title > a > strong')
    print(title)
    day = soup.select('div.author')
    for i in day:
        a = i.text.replace('\t','').replace('\n','')
        print(a)

    #print(title)
    #day = day.text.split('|')[4].replace('\t','').replace('\n','')
    # 결과는 리스트
    
    bt = []
    bd = []

    for i in title:
        b_title = i.text
        print(b_title)
        bt.append(b_title)    

    for k in a:
        print(k)
        bd.append(k)
    
pageNum = pageNum + 1
```

```python
gyobo = pd.DataFrame({'책이름' : bt,
                      '발행일자' : bd}
)
gyobo
```
