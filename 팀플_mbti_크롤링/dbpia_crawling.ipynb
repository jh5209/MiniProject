### dbpia 논문 사이트 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
import datetime
import csv
```

```python
# 브라우저 생성
browser = webdriver.Chrome('D:/VS_work/crawling/chromedriver.exe')

# 웹사이트 열기
browser.get('https://www.dbpia.co.kr/')

# 웹사이트 로딩이 끝날 때까지 10초 동안 기다려 줌
browser.implicitly_wait(10)

# 검색창 클릭
search2 = browser.find_element_by_id('keyword')
search2.send_keys('mbti')
time.sleep(2)

# 검색돋보기 클릭
search3 = browser.find_element_by_xpath('//*[@id="bnHead"]/div[3]/div/div[1]/div[1]/a')
search3.click()
browser.implicitly_wait(10)
```

```python
# 제목, 발행날짜 추출하기
# 제목, 발행날짜 리스트 생성
b_title = []
b_day = []

items = browser.find_elements_by_css_selector('.listBox')

# csv 파일 생성
f = open(f'{mbti}.csv', 'w', encoding='CP949', nemline = '')
csvWriter = csv.writer(f)

for item in items:
    
    try:
        title = item.find_element_by_class_name('titWrap').text
        day = item.find_element_by_class_name('date').text
        print(title, day)
        b_title.append(title)
        b_day.append(day)
        
        search4 = browser.find_element_by_class_name('viewMore')
        search4.click()
        time.sleep(2)

    except: # 예외처리 
        pass

# csv 파일에 넣기
for i in b_title:
    csvWriter.writerow(i)

for j in b_day:
    csvWriter.writerow(j)

# csv 파일 닫기
f.close()

        
```
