##### 수도권매립지관리공사에 반입되는 서울, 경기도, 인천 생활폐기물 반입량 통계
자료출처: https://gimi9.com/organization/data-go-kr

수도권매립지 : 인천광역시 서구 검단동 및 검암동, 경기도 김포시 양촌면 일원 <br>
1~4매립장 현재 1,2 매립장은 매립완료이며 3,4 매립장 사용 중이다.

```python
# 수도권매립지관리공사에 들어오는 수도권 생활폐기물 반입량

df_b = pd.read_csv('./data/수도권매립지관리공사_서울시_연도별 _폐기물_반입량.csv', encoding='CP949')
df_b1 = df_b.copy()
df_b1
```

```python
df_c = pd.read_csv('./data/수도권매립지관리공사_경기도_연도별_폐기물_반입량.csv', encoding='CP949')
df_c1 = df_c.copy()
df_c1 
```

```python
df_i = pd.read_csv('./data/수도권매립지관리공사_인천시_연도별_폐기물_반입량.csv', encoding='CP949')
df_i1 = df_i.copy()
df_i1 
```

```python
df_bci =pd.concat([df_b1,df_c1,df_i1], ignore_index=True)
df_bci
```

```python
# year 행값 추출
year = []
for i in df_bci.iloc[0:22, 0]:
    year.append(i)
year
```

```python
# seoul 행값 추출
seoul = []
for i in df_bci.iloc[0:22, -1]:
    seoul.append(i)
seoul
```

```python
# incheon 
incheon = []
for i in df_bci.iloc[44:67, -1]:
    incheon.append(i)
incheon
```

```python
# Gyeonggi
gyeonggi = []
for i in df_bci.iloc[22:44, -1]:
    gyeonggi.append(i)
gyeonggi
```

```python

# 수도권매립지에서 반입하는 수도권의 지자체별 생활폐기물 반입량 그래프

plt.rc('font', family='Malgun Gothic')
fig = plt.figure(figsize=(15, 8)) # 그래프 크기 키우기
x_axis = year
# y_axis = [0, 10*100, 15000000]

data1 = seoul
data2 = gyeonggi
data3 = incheon

CIRCLE_MARKER = 'o'
SQUARE_MARKER = 's'
TRIANGLE_MARKER = '^'

plt.title('수도권 지자체별 생활폐기물 반입량')
plt.plot(x_axis, data1, marker = CIRCLE_MARKER, label='서울시')
plt.plot(x_axis, data2, marker = SQUARE_MARKER, label='경기도')
plt.plot(x_axis, data3, marker = TRIANGLE_MARKER, label='인천시')
plt.legend(loc=5)
plt.xticks(x_axis)
# plt.yticks(y_axis)
plt.show()
```
